# Frequency Domain Green's Function — Reproduction

This notebook reproduces the figures using the saved arrays in `FrequencyDomainGreensFunction_plot_arrays.h5`.

```julia
using HDF5
using Plots
using LaTeXStrings

h5file = joinpath(@__DIR__, "FrequencyDomainGreensFunction_plot_arrays.h5")

PIA_Freqs = h5read(h5file, "raw/PIA_Freqs")
NIA_Freqs_plot = h5read(h5file, "raw/NIA_Freqs_plot")
G_plus_PIA = h5read(h5file, "raw/G_plus_PIA")
G_minus_NIA_plot = h5read(h5file, "raw/G_minus_NIA_plot")
arc_Freqs = h5read(h5file, "raw/arc_Freqs")
G_minus_arc = h5read(h5file, "raw/G_minus_arc")

x_pia = h5read(h5file, "plot/p1/x_pia")
y_pia_re = h5read(h5file, "plot/p1/y_pia_re")
y_pia_im = h5read(h5file, "plot/p1/y_pia_im")
x_nia = h5read(h5file, "plot/p1/x_nia")
y_nia_re = h5read(h5file, "plot/p1/y_nia_re")
y_nia_im = h5read(h5file, "plot/p1/y_nia_im")

x_arc = h5read(h5file, "plot/p3/x_arc")
y_arc_re = h5read(h5file, "plot/p3/y_arc_re")
y_arc_im = h5read(h5file, "plot/p3/y_arc_im")

println("Loaded: ", h5file)
```

## Figure 1 — Frequency domain on PIA and NIA

```julia
p1 = plot(x_pia, y_pia_re, label = L"\Re\, [G^+(\omega)]", lw = 2)
plot!(p1, x_pia, y_pia_im, label = L"\Im\, [G^+(\omega)]", lw = 2)
plot!(p1, x_nia, y_nia_re, label = L"\Re\, [G^-(\omega)]", lw = 2)
plot!(p1, x_nia, y_nia_im, label = L"\Im\, [G^-(\omega)]", lw = 2)
plot!(p1, frame = :box, title = "", xlabel = L"\Im \, (\omega)", legend = :topright, yscale = :log10, figsize = (300, 200))
savefig(joinpath(@__DIR__, "Frequency_Domain_FG_PIA_NIA.pdf"))
p1
```

## Figure 2 — Small arc

```julia
p3 = plot(x_arc, y_arc_re, label = L"\Re \, [G^-]", lw = 2)
plot!(p3, x_arc, y_arc_im, label = L"\Im \, [G^-]", lw = 2)
plot!(p3, frame = :box, title = "", xlabel = L"\phi", ylabel = "", figsize = (300, 200))
savefig(joinpath(@__DIR__, "Frequency_Domain_FG_on_Small_Arc.pdf"))
p3
```
